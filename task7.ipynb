```python
total_amount=0
item_count=0
print("enter the product price:")
while True:
  price=float(input())
  if price==0:
    break
  if price<0:
    print("no negative price allowed:")
    continue
  total_amount+=price
  item_count+=1

  if total_amount>=1000:
    discount=0.20
  elif total_amount>=500:
    discount=0.10
  elif total_amount>=200:
    discount=0.05
  else:
    discount=0

  total_discount=total_amount*discount # Corrected: discount is a percentage, so multiply by total_amount
  final_amount=total_amount-total_discount # Corrected: final amount should subtract discount
  print("total items:",item_count)
  print("subtotal:",total_amount)
  print("discount amount:",total_discount)
  print("final total:",final_amount)
```
